```python
labels = ['KC1', 'KC2', 'JM1', 'PC1', 'PC3', 'PC4']
bal_NN_SMOTE = [70.54, 75.42, 64.03, 71.48, 76.69, 80.86]
bal_NN = [42.28, 61.86, 35.74, 43.17, 41.81, 61.31]

x = np.arange(len(labels))  # the label locations
width = 0.40  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, bal_NN_SMOTE, width, label='NN+SMOTE')
rects2 = ax.bar(x + width/2, bal_NN, width, label='NN')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Bal')
ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(top=90)
ax.legend(loc=(0,0))


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(3, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.savefig('bal_NN_SMOTE.png')

plt.show()
```
